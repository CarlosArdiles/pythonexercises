```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import networkx as nx
import pandas as pd
```

```python
#1. LINE PLOT
dict = {
    'Enero':1550,
    'Febrero':2800,
    'Marzo':200,
    'Abril':1000,
    'Mayo':1700,
    'Junio':3200,
}
meses = list(dict.keys())
utilidad = list(dict.values())

sns.lineplot(x=meses, y =utilidad)
plt.title("Utilidad por mes")
plt.xlabel("meses")
plt.ylabel("Utilildades")
plt.show()
```

El gráfico de línea se utiliza para visualizar la relación y la tendencia de una variable numérica a lo largo de un eje de tiempo o un eje continuo. Se utiliza comúnmente para representar series de tiempo, donde el eje x representa unidades de tiempo, como días, meses o años, y el eje y representa los valores de la variable en estudio.

El gráfico de línea se utiliza para:

1. Mostrar la evolución y la tendencia de una variable a lo largo del tiempo.
2. Identificar patrones y fluctuaciones en los datos a lo largo del tiempo.
3. Comparar múltiples series de tiempo en un mismo gráfico.
4. Identificar puntos de inflexión, cambios o tendencias ascendentes o descendentes en los datos.
5. Realizar pronósticos o proyecciones sobre el comportamiento futuro de la variable.

El objetivo principal del gráfico de línea es proporcionar una representación visual clara de la evolución de una variable a lo largo del tiempo. Permite observar patrones, identificar cambios significativos y realizar análisis de tendencias. Es ampliamente utilizado en áreas como la economía, las finanzas, la meteorología, el análisis de ventas, la investigación científica y otros campos donde se requiere el análisis de series de tiempo y datos continuos.

```python
#2. Scatter Plot
data = {'precio': [250000, 300000, 350000, 400000, 450000],
        'tamano': [150, 180, 200, 220, 240]}
df = pd.DataFrame(data)
sns.scatterplot(data = df , x = 'tamano' , y = 'precio')
plt.title('Relación entre el precio y el tamaño de las viviendas')
plt.xlabel('Tamaño (m^2)')
plt.ylabel('Precio (USD)')
plt.show()
```

Un scatter plot utiliza dos variables numéricas, una en el eje x y otra en el eje y. Sirve para visualizar la relación o la correlación entre estas dos variables.

El objetivo principal de un scatter plot es mostrar cómo se distribuyen los puntos en el gráfico en función de las dos variables. Esto ayuda a identificar patrones, tendencias o relaciones entre las variables. Al visualizar la dispersión de los puntos en el gráfico, se puede determinar si existe una relación lineal, una relación no lineal, una correlación positiva, una correlación negativa o ninguna correlación entre las variables.

El scatter plot es útil en diversos contextos, como en análisis exploratorios de datos, modelado de relaciones entre variables, identificación de patrones o anomalías, y análisis de tendencias. Además, se puede utilizar para comparar datos de diferentes grupos o categorías, agregando colores o formas diferentes a los puntos según una variable categórica adicional.

```python
#3. BAR PLOT
dict = {
    'BMW' : 40,
    'CHEVROTEL' : 10,
    'AUDIO' : 35,
    'FORD' : 40,
}
categories = list(dict.keys())
values = list(dict.values())
sns.barplot(x=categories, y=values, hue=categories)
plt.title("Gráfico Ventas por Marca")
plt.xlabel("Marcas de autos")
plt.ylabel("Precio promedio en Millones de pesos")
plt.show()

```

El gráfico de barras se utiliza para visualizar la distribución de una variable categórica o discreta. Este tipo de gráfico representa cada categoría como una barra, donde la altura de la barra representa la frecuencia o el conteo de esa categoría en los datos.

El gráfico de barras se utiliza para:

1. Comparar la frecuencia o el conteo de diferentes categorías.
2. Identificar las categorías más comunes o frecuentes.
3. Mostrar la distribución de una variable categórica.
4. Realizar análisis de frecuencia o conteo.

El objetivo principal del gráfico de barras es proporcionar una representación visual clara de la distribución de una variable categórica. Permite identificar rápidamente las categorías más frecuentes o dominantes, así como comparar las frecuencias o conteos entre diferentes categorías. Es ampliamente utilizado en áreas como el análisis de encuestas, el estudio de preferencias de los consumidores, el análisis de datos demográficos y en general en cualquier situación donde se requiera analizar la distribución de una variable categórica.

```python
# 4 HISTOGRAM CHART

# Crear un conjunto de datos de ejemplo
data = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70]

# Crear el histograma con Seaborn
sns.histplot(data)

# Personalizar el gráfico
plt.title('Histograma de datos')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()
```

El histograma es una representación gráfica que se utiliza para visualizar la distribución de una variable numérica. El eje x del histograma representa los diferentes valores de la variable, divididos en intervalos, y el eje y muestra la frecuencia con la que esos valores ocurren en los datos.

El histograma se utiliza para:

1. Visualizar la forma y la estructura de la distribución de una variable.
2. Identificar patrones, como la presencia de picos, asimetrías o tendencias.
3. Detectar la presencia de valores atípicos o valores extremos.
4. Comparar la distribución de diferentes variables o subgrupos de datos.

El objetivo principal del histograma es proporcionar una representación visual de la distribución de los datos, lo que ayuda a comprender mejor la variabilidad y las características de la variable en estudio. Además, el histograma puede ser útil para realizar análisis exploratorios de datos y tomar decisiones informadas en diferentes áreas, como la estadística, la investigación científica, el análisis de datos, entre otros.

```python
#5. PIE CHART

# Datos de ejemplo Ventas en miles de millones de USD
marcas = ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Otras']
ventas = [45, 30, 15, 8, 2]

# Configuración del gráfico de torta
colors = sns.color_palette('Set3')
explode = (0.05, 0, 0, 0, 0)  # Para resaltar una porción del gráfico

# Crear el gráfico de torta
plt.pie(ventas, explode=explode, labels=marcas, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Para que el gráfico sea circular

# Agregar título
plt.title('Ventas de teléfonos por marca')

# Mostrar el gráfico
plt.show()
```

Un gráfico de torta (pie chart) se utiliza comúnmente para representar la distribución de una variable categórica en un conjunto de datos. Es decir, muestra cómo se dividen proporcionalmente las diferentes categorías en relación con el total.

El objetivo principal de un gráfico de torta es resaltar la proporción de cada categoría en relación con el conjunto de datos completo. Es especialmente útil cuando se quiere visualizar la contribución relativa de cada categoría y comparar su tamaño en relación con el todo.

Un gráfico de torta es adecuado para variables categóricas nominales o ordinales, donde no hay un orden específico entre las categorías. Se muestra mediante sectores circulares que representan las proporciones de cada categoría en función de su valor absoluto o porcentaje.

Este tipo de gráfico es ampliamente utilizado en informes, presentaciones y análisis de datos para proporcionar una representación visual clara de la distribución de una variable categórica y resaltar las categorías más grandes o importantes en relación con las demás.

```python
#6 BOX PLOT
# Datos de ejemplo
datos = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75]

sns.boxplot(data=datos)

plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Boxplot')
plt.show()
```

Un boxplot se utiliza para representar la distribución y la variabilidad de un conjunto de datos numéricos. Se utiliza comúnmente para visualizar la distribución de una variable cuantitativa y para identificar posibles valores atípicos o valores extremos.

El boxplot utiliza una caja (box) que representa el rango intercuartílico (el rango que abarca desde el primer cuartil al tercer cuartil), una línea en medio de la caja que representa la mediana, y "bigotes" que se extienden desde la caja hasta los valores mínimo y máximo dentro de un rango específico.

El boxplot permite identificar visualmente la simetría de la distribución, la presencia de valores atípicos y la dispersión de los datos. También puede comparar varias distribuciones utilizando múltiples boxplots en el mismo gráfico.

En términos de tipo de variable, un boxplot se utiliza principalmente para variables numéricas continuas o variables ordinales. Sin embargo, también se pueden crear boxplots para variables categóricas si se asignan valores numéricos a las categorías.


```python
#7 VIOLIN PLOT
# Ejemplo de datos
data = sns.load_dataset("iris")

# Crear el violin plot
sns.violinplot(x="species", y="sepal_length", data=data)


plt.xlabel("Especies")
plt.ylabel("Longitud del sépalo")
plt.title("Distribución de la longitud del sépalo por especie")
plt.show()
```

Un violin plot es útil para visualizar la distribución de una variable numérica en diferentes categorías o grupos. Se utiliza principalmente para analizar la forma y la variabilidad de los datos en cada categoría.

El violin plot utiliza una combinación de un box plot y un kernel density plot. En el eje y, muestra la densidad de probabilidad estimada de los valores de la variable, lo que permite identificar la forma y la prominencia de los picos en la distribución. Además, muestra un box plot que representa la mediana, los cuartiles y los valores atípicos de los datos.

El violin plot es especialmente útil cuando se desea comparar la distribución de una variable en diferentes grupos o categorías. Puede revelar diferencias en la forma, la concentración y la dispersión de los datos entre las categorías. También permite identificar valores atípicos y evaluar la simetría o asimetría de la distribución.

En resumen, el violin plot es una herramienta poderosa para visualizar y comparar distribuciones de variables en diferentes grupos, lo que facilita la exploración y el análisis de datos.

```python
#8 HEATMAPS

data = pd.DataFrame(np.random.rand(10, 10))

# Crear el heatmap
sns.heatmap(data, cmap='coolwarm', annot=True)
plt.show()
```

Un heatmap (mapa de calor) utiliza principalmente variables numéricas continuas para representar la intensidad de un fenómeno o valor en una matriz o tabla. Cada celda del heatmap está coloreada según el valor numérico correspondiente, lo que permite visualizar patrones, tendencias y relaciones entre las variables.

El heatmap es útil para:

1. Identificar patrones y estructuras en datos tabulares: El heatmap muestra visualmente las diferencias en los valores de las celdas, lo que permite identificar patrones, tendencias o grupos en los datos.

2. Visualizar correlaciones: En el caso de una matriz de correlación, el heatmap muestra la fuerza y dirección de las relaciones entre variables, resaltando las correlaciones positivas y negativas.

3. Destacar valores atípicos o anomalías: Un heatmap puede ayudar a identificar valores inusuales o atípicos en una matriz o tabla de datos, ya que estos valores se destacarán con colores diferentes a los demás.

4. Comparar distribuciones: El heatmap puede utilizarse para comparar la distribución de valores en diferentes grupos o categorías, lo que permite identificar diferencias significativas entre ellos.

```python
#9 3D PLOTS
# Datos de ejemplo
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Crear la figura y los ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear el gráfico de dispersión en 3D
ax.scatter(x, y, z)

# Configurar etiquetas de los ejes
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
```

Un gráfico en 3D se utiliza generalmente para visualizar y representar datos que tienen tres dimensiones. Esto significa que se utilizan tres variables para representar los datos en los ejes x, y y z.

El objetivo principal de un gráfico en 3D es mostrar la relación entre tres variables y cómo varían conjuntamente. Al visualizar los datos en un espacio tridimensional, se pueden identificar patrones, tendencias o relaciones complejas que podrían ser difíciles de percibir en gráficos bidimensionales.

Los gráficos en 3D son útiles en diversas áreas, como la visualización de datos científicos, el análisis de modelos matemáticos o físicos, la representación de superficies tridimensionales, entre otros. Ayudan a comprender mejor la estructura de los datos y facilitan la identificación de patrones espaciales o tendencias en conjuntos de datos complejos.

```python
#10 POLAR PLOT

theta = np.linspace(0, 2*np.pi, 100)
r = np.random.rand(100)

# Crear el gráfico polar
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(theta, r)

# Personalizar el gráfico
ax.set_rticks([])
ax.set_yticklabels([])

# Mostrar el gráfico
plt.show()
```

Un polar plot, también conocido como gráfico polar o gráfico de coordenadas polares, se utiliza para visualizar datos que tienen una relación con una variable angular y una variable radial. 

En un polar plot, la variable angular se representa en el eje circular, mientras que la variable radial se representa en el radio o distancia desde el centro del gráfico. El objetivo principal de un polar plot es mostrar patrones, tendencias o distribuciones de datos en función de su ángulo y magnitud.

Este tipo de gráfico es útil en diversas áreas, como matemáticas, física, ingeniería y ciencias sociales. Algunos ejemplos comunes de uso de un polar plot incluyen representar coordenadas geográficas, análisis de ondas y oscilaciones, análisis de datos circulares (por ejemplo, dirección del viento) y visualización de datos en coordenadas polares.

En resumen, el polar plot es una herramienta efectiva para representar datos que tienen una relación angular y radial, permitiendo una visualización clara de patrones y tendencias en datos circulares.

```python
# 11 NETWORK GRAPH
# Crear un objeto de grafo
G = nx.Graph()

# Agregar nodos al grafo
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")

# Agregar aristas al grafo
G.add_edge("A", "B")
G.add_edge("B", "C")
G.add_edge("C", "D")
G.add_edge("D", "A")

# Dibujar el gráfico de red
nx.draw(G, with_labels=True, node_color='skyblue', edge_color='gray')

# Mostrar el gráfico
plt.show()
```

Un gráfico de red (network graph) se utiliza para visualizar y analizar relaciones o conexiones entre entidades. En este tipo de gráfico, las entidades se representan como nodos y las relaciones entre ellas se representan como aristas.

El tipo de variable utilizado en un gráfico de red es un conjunto de datos estructurados que define las entidades y las conexiones entre ellas. Por lo general, se utilizan estructuras de datos como listas de adyacencia, matrices de adyacencia o archivos de formato específico (por ejemplo, GML, GraphML) para representar la estructura del grafo.

El objetivo principal de un gráfico de red es visualizar y comprender la estructura y las relaciones entre las entidades en un sistema o conjunto de datos. Puede ser utilizado en diversas áreas, como ciencias sociales, análisis de redes, biología, transporte, informática, entre otros. Al visualizar las conexiones entre entidades, los gráficos de red pueden ayudar a identificar patrones, comunidades, centralidad de nodos, rutas más cortas, flujos de información, entre otros análisis de redes.

```python
#12. ERROR BAR
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 5, 6, 7])
error = np.array([0.5, 0.3, 0.7, 0.2, 0.4])

# Crear el gráfico
plt.errorbar(x, y, yerr=error, fmt='o', capsize=4)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Barras de Error')
plt.show()
```

Un gráfico de barras de error (error bar) se utiliza para visualizar la incertidumbre o variabilidad en los datos. Se utiliza para representar la dispersión de los datos y mostrar la confiabilidad de las mediciones.

En un gráfico de barras de error, se representa la medida central (como la media) con una barra vertical y se muestran los intervalos de confianza o desviación estándar alrededor de esta medida central. Esto permite comparar diferentes grupos de datos y evaluar si las diferencias observadas son estadísticamente significativas.


```python
# 13 . Subplots and multiple axes 

# Crear los datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x)

# Crear una figura con subplots y múltiples ejes
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Subplot 1: Gráfico de línea
axes[0, 0].plot(x, y1, color='blue')
axes[0, 0].set_title('Función Seno')

# Subplot 2: Scatter plot
axes[0, 1].scatter(x, y2, color='red')
axes[0, 1].set_title('Scatter Plot')

# Subplot 3: Gráfico de barras
axes[1, 0].bar(x, y3, color='green')
axes[1, 0].set_title('Gráfico de Barras')

# Subplot 4: Histograma
axes[1, 1].hist(y1, bins=20, color='purple')
axes[1, 1].set_title('Histograma')

# Ajustar los espacios entre los subplots
plt.tight_layout()

# Mostrar la figura
plt.show()
```

Un Subplots and Multiple Axes, también conocido como subgráficos o gráficos múltiples, utiliza variables de tipo figura (Figure) y ejes (Axes) en la librería Matplotlib.

El objetivo principal de los subplots y múltiples ejes es mostrar múltiples gráficos dentro de una misma figura de manera organizada. Esto es útil cuando se desea comparar diferentes conjuntos de datos o visualizar diferentes aspectos de un conjunto de datos en un solo lugar.

Las variables de tipo figura (Figure) representan la ventana o lienzo en el cual se dibujan los gráficos. Los subplots se crean dentro de una figura y se organizan en una matriz de filas y columnas.

Las variables de tipo ejes (Axes) representan un sistema de coordenadas en el cual se dibujan los gráficos individuales. Cada eje se ajusta dentro de un subplot y se puede personalizar con diferentes tipos de gráficos, estilos, colores, etiquetas, entre otros.

El uso de subplots y múltiples ejes permite mostrar y comparar fácilmente múltiples gráficos en una misma figura, lo cual es especialmente útil en análisis exploratorio de datos, presentación de resultados y visualización de información compleja.
