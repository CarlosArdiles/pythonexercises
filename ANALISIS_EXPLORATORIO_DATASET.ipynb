El supuesto a estudiar es verificar si es causal que a mayor indice de acceso a carreteras implica que la vivienda tenga un mayor pago de impuestos.

```python
df1 = df[['RAD', 'TAX']]
```

pensaba eliminar los puntos que graficamente parecian atipicos, pero viendo el listado son 132 datos que tienen esta caracteristica, esto se puede interpretar como la gran brecha del poder adquisitivo que existe en esta zona.

```python
import statsmodels.formula.api as smf
```

```python
df1.describe()
```

```python
lm = smf.ols(data=df, formula='TAX ~ RAD').fit()
```

Nuestro modelo tiene un tipo de regresión lineal.


```python
lm.params
```

```python
lm.params[0]
```

```python
lm.params[1]
```

```python
lm.summary()
```

Conclusiones de la data estudiada para la regresión:  

El impuesto más alto y más bajo a pagar es de : 711 y 187 respectivamente.  
Índice de accesibilidad a carreteras radiales max y min:24 y 1 respectivamente.  
Hay una gran brecha en la paga de impuestos, pero una baja diferencia en el  Índice de accesibilidad a carreteras radiales.  
Puedo inferir que existe un pago excesivo de impuestos por un bajo o dificultoso acceso a una carretera.


Conclusiones Summary:

de R podemos indicar que la variable que nos indica un indice de acceso a las carreteras afecta en un 80% a la variable del impuesto que debe pagar dicha vivienda.

con Prob(F) podemos verificar que es inferior a 0.01 por lo que podemos decir que esta relación es real.

con Log-likelihood podemos comprobar que la data no fue generada por el modelo.

en el 95% de los casos estará en el rango de 16.917 y 18.320, esto tambien nos indica que es bastante bajo el indice de acceso a las carreteras en la zona de estudio.

¿Cuáles son las variables que más inciden en el precio de una propiedad? 
 

Estudio multivariable:

```python
df2= df[['MEDV', 'RM', 'LSTAT','PTRATIO']]
```

```python
df2.head()
```

```python
sns.pairplot(data=df2)
```

```python
df2.corr()
```

```python
sns.heatmap(df2.corr(), cmap='coolwarm', annot=True)
```

```python
lm2 = smf.ols(formula='''Q('MEDV') 
                            ~ Q("RM") + 
                              Q("LSTAT") + Q('PTRATIO')''',data=df2).fit()
```

```python
lm2.params
```

```python
lm2.summary()
```

Podemos evidenciar que ambas variables RM , LSTAT , PTRATIO afectan en un 67.7% al precio de la vivienda.  
tenemos un valor de P muy inferior a 0.01 por lo que podemos concluir que estas variables tienen una relacion real con el precio de la vivienda.  
también al analizar el summary determinamos que las variables estudiadas no afectan por casualidad al precio de la vivienda.
